```python
# Libraries
import ipynb.fs.full.market_ingestion_coinmarketcap
from ipynb.fs.full.market_ingestion_coinmarketcap import MarketListener
import ipynb.fs.full.market_register_coinmarketcap
from ipynb.fs.full.market_register_coinmarketcap import MarketRegister
import ipynb.fs.full.market_processing_coinmarketcap
import ipynb.fs.full.market_register_coinmarketcap_s
from ipynb.fs.full.market_register_coinmarketcap_s import SummaryMarketRegister
from ipynb.fs.full.market_processing_coinmarketcap import MarketProcessing
import time
from coinmarketcap import Market
```

```python
if __name__ == '__main__':
    
    api_market = Market()
    
    start_time = time.time()
    time_limit = 180 # define the execution time
    i = 1
    
    """
    Start to ingest the data in batch/streaming files ...
    """                                       
    print("* Start to ingest the data in batch/streaming files ...") 
    
    while True: 
        
        start_api_market = time.time() 
        
        """
            Call the class MarketListener
        """        
        market_listener = MarketListener(api=api_market,time_limit=60,batch_market_filename='raw_market_batch.json',
                                         streaming_market_filename='raw_market_streaming.json',
                                         batch_summary_market_filename='raw_summary_market_batch.json',
                                         streaming_summary_market_filename='raw_summary_market_streaming.json')
        market = market_listener.streaming()
        
        """
            Call the class MarketRegister & SummaryMarketRegister
        """
        register = MarketRegister(filename='raw_market_streaming.json')
        market = register.build_data()
        
        register = SummaryMarketRegister(filename='raw_summary_market_streaming.json')
        summary_market = register.build_data()
        
        """
            Start to build the data from streaming file ...
        """                                       
        print("** Step {0} of the data building from streaming file ...".format(i)) 
        
        """
            Call the class MarketProcessing
        """
        Processing = MarketProcessing([market,summary_market])
        data_market = Processing.featurize()
    
        # export the data to csv format file
        Processing.put_csv(data=data_market,filename='market_streaming_btc.csv',type_mode='w+')
        
        end_api_market = time.time()
        runtime_api_market = end_api_market - start_api_market
        
        print("-- runtime api market : {0}".format(runtime_api_market))
        
        i=i+1
        
        if (time.time() - start_time) > time_limit:
            break
         
    """
        Call the class MarketRegister & SummaryMarketRegister
    """      
            
    register = MarketRegister(filename='raw_market_batch.json')
    market = register.build_data()
    
    register = SummaryMarketRegister(filename='raw_summary_market_batch.json')
    summary_market = register.build_data()
    
    """
    Start to build the data from batch file ...
    """                                       
    print("* Step of the data building from batch file ...")
    
    """
        Call the class MarketProcessing & SummaryMarketRegister
    """
    Processing = MarketProcessing([market,summary_market])
    data_market = Processing.featurize()    
    
    # export the data in csv format file
    Processing.put_csv(data=data_market,filename='market_batch_btc.csv',type_mode='w')
    
    print('* Finish !')
    
```
